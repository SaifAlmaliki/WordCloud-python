```
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```
text = '''Data science is an inter-disciplinary field that uses scientific methods, processes, algorithms 
and systems to extract knowledge and insights from many structural and unstructured data.
Data science is related to data mining, machine learning and big data.
Data science is a "concept to unify statistics, data analysis and their related methods" in order to "understand 
and analyze actual phenomena" with data. It uses techniques and theories drawn from many fields 
within the context of mathematics, statistics, computer science, domain knowledge and information science.'''
```

Create and Generate wordcloud Image from the text

```
wc = WordCloud()
wc.generate(text)

# Display the general image
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
wc.to_file('first_review.png')
```

Remove **stopwords** from the sentences

```
stopwords=set(STOPWORDS)
stopwords.add('Data')                       # consider 'data' as stopword
stopwords.update('methods', 'see', 'seen')  # filter these words as stopwords

# Display the general image
wc = WordCloud(stopwords=stopwords, max_words=1000, background_color='white')
wc.generate(text)

plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

**Pull text from Wikepedia**

```
!pip install wikipedia
import wikipedia

page = wikipedia.page('Natural Language Processing')
text =  page.content
wc = WordCloud().generate(text)

# Adjust the image size
plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=2)
plt.show()
```

Customize the word width, height. background color

```
wc = WordCloud(width=800,
               height = 600,
               min_font_size=14,
               background_color="#333333",
               colormap="spring").generate(text)

# Adjust the image size
plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=2)
plt.show()
```

**Mask Example**

Add image as mask to your word

```
from PIL import Image
import numpy as np

image = Image.open("heart.jpg")
mask = np.array(image)
wc = WordCloud(mask=mask, colormap="Reds").generate(text)

plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=2)
plt.show()
```

**Image Color Generator**

Add colored image as mask

```
from wordcloud import ImageColorGenerator

image = Image.open("color_heart.jpg")
mask = np.array(image)

# Add color to my image
color_image = ImageColorGenerator(mask)

wc = WordCloud(mask=mask, color_func=color_image).generate(text)


plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=2)
plt.show()
```

**Custom Color Function**

any word has 't' letter will be yellow

```
def custom_color(*args, **kwrgs):
    word = args[0]
    if 't' in word:
        return "#FFD700"
    return "#CCCCCC"

wc = WordCloud(color_func=custom_color).generate(text)

plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=2)
plt.show()
```

**Generate from Frequecies**

show the most frequent words, not the most important

```
from collections import Counter

page = wikipedia.page('Internet Of Things')
text =  page.content
tokens = text.split(" ")
word_count = Counter(tokens)   # count the frequency of each word

wc = WordCloud().generate(text)
wc.generate_from_frequencies(word_count)  # Most frequent , not important

plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=2)
plt.show()
```

**Mask Equation**

```
x, y =np.ogrid[:300, :300]
mask =(x - 150)**2 + (y-150)**2 > 130**2
mask =255* mask.astype(int)

wc= WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)


plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
